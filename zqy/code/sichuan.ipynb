# PCA

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from torchvision import transforms


def pca_lowrank(input: pd.DataFrame, proportion: float) -> torch.tensor:
    # Convert pandas DataFrame to torch tensor
    input_tensor = torch.tensor(input.values, dtype=torch.float32)

    # Normalize the data
    mean = torch.mean(input_tensor, dim=0)
    std = torch.std(input_tensor, dim=0)
    std[std == 0] = 1e-3 # Avoid division by zero
    input_normalized = (input_tensor - mean) / std

    # Calculate the number of principal components based on the proportion of variance to retain
    U, S, V = torch.pca_lowrank(input_normalized)
    total_variance = torch.sum(S**2)
    variance_retained = 0.0
    num_components = 0

    for i in range(len(S)):
        variance_retained += S[i]**2
        num_components += 1
        if variance_retained / total_variance >= proportion:
            break

    # Apply PCA with the calculated number of components
    train_data_pca = torch.matmul(input_normalized, V[:, :num_components])

    return train_data_pca
```

```python
train_data_pca = pca_lowrank(train_data_fillna, 0.99)
train_data_pca.shape
```

```python
# Assuming batch_num, batch_size, and len are defined
batch_size = 50  # Example value
batch_num = train_data_pca.shape[0] // batch_size
length = train_data_pca.shape[1]  # This should be the same as the number of principal components

# Reshape train_data_pca
train_data_pca_reshaped = train_data_pca[:batch_size*batch_num].view(batch_num, batch_size, length)
train_data_pca_reshaped.shape
```

# MLP

```python
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size, bias=True)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size, bias=True)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(output_size, 1, bias=True)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x
```

```python
from sklearn.metrics import f1_score, auc, roc_curve


class Train:
    def __init__(self, model, criterion, optimizer, batch_size):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.batch_size = batch_size

    def train(self, train_data, train_target, epochs):

        for epoch in range(epochs):
            loss_total = 0
            accuracy_total = 0
            total = 0
            all_preds = np.array([])
            all_labels = np.array([])
            for i in range(train_data.shape[0]):
                self.optimizer.zero_grad()
                output = self.model(train_data[i])
                loss = self.criterion(output, train_target[i])
                loss.backward()
                self.optimizer.step()
                # compute the accuracy
                loss_total += loss.item()
                total += train_target[i].shape[0]
                accuracy_total += torch.sum(
                    (
                        (torch.round(output.detach()) - train_target[i]).abs() < 1e-3
                    ).int()
                ).item()
                # compute the F1 score, AUC, and ROC curve
                all_preds = np.concatenate(
                    [all_preds, torch.round(output.detach()).numpy().flatten()]
                )
                all_labels = np.concatenate([all_labels, train_target[i].numpy().flatten()])
            print(
                f"Epoch {epoch + 1}, Loss: {loss_total}, Accuracy: {accuracy_total / total}," +
                f"F1 Score: {f1_score(all_labels, all_preds)}"
            )
```

```python
mlp = MLP(input_size=train_data_pca.shape[1], hidden_size=32, output_size=16)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(mlp.parameters(), lr=0.01)
trainer = Train(model=mlp, criterion=criterion, optimizer=optimizer, batch_size=batch_size)
trainer.train(train_data_pca_reshaped, train_target_drop, epochs=10)
```
